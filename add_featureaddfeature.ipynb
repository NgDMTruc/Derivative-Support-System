```python
import pandas as pd
import numpy as np
```

```python
def data_backtest(data):
    new_data = data[['Open', 'High', 'Low', 'Close', 'Volume']]
    new_data.index = pd.to_datetime(data['Date'])
    new_data.index.name = None
    return new_data
```

```python
df_min = pd.read_csv('vn30f1m_3min.csv')
df_hour = pd.read_csv('vn30f1m_1hour.csv')
df_day = pd.read_csv('vn30f1m_1day.csv')

data_min = data_backtest(df_min)
data_hour = data_backtest(df_hour)
data_day = data_backtest(df_day)
```

```python
import pandas as pd

def label_quarter(df):
    # Lấy cột index dạng datetime
    df = df.copy()
    df['date'] = df.index

    # Tính ngày thứ 21 của tháng đầu tiên trong mỗi quý
    def get_quarter_label(date):
        if date.month in [1, 2, 3]:  # Quý 1 (Bắt đầu từ 21/01)
            start_of_quarter = pd.Timestamp(year=date.year, month=1, day=21)
            if date >= start_of_quarter:
                return f"{date.year}Q1"
            else:
                return f"{date.year - 1}Q4"  # Nếu trước 21/01, thuộc quý 4 của năm trước

        elif date.month in [4, 5, 6]:  # Quý 2 (Bắt đầu từ 21/04)
            start_of_quarter = pd.Timestamp(year=date.year, month=4, day=21)
            if date >= start_of_quarter:
                return f"{date.year}Q2"
            else:
                return f"{date.year}Q1"  # Nếu trước 21/04, thuộc quý 1

        elif date.month in [7, 8, 9]:  # Quý 3 (Bắt đầu từ 21/07)
            start_of_quarter = pd.Timestamp(year=date.year, month=7, day=21)
            if date >= start_of_quarter:
                return f"{date.year}Q3"
            else:
                return f"{date.year}Q2"  # Nếu trước 21/07, thuộc quý 2

        elif date.month in [10, 11, 12]:  # Quý 4 (Bắt đầu từ 21/10)
            start_of_quarter = pd.Timestamp(year=date.year, month=10, day=21)
            if date >= start_of_quarter:
                return f"{date.year}Q4"
            else:
                return f"{date.year}Q3"  # Nếu trước 21/10, thuộc quý 3

    # Áp dụng hàm get_quarter_label cho từng hàng của DataFrame
    df['quarter_label'] = df['date'].apply(get_quarter_label)

    return df['quarter_label']


data_min['quarter_label'] = label_quarter(data_min)
data_hour['quarter_label'] = label_quarter(data_hour)
data_day['quarter_label'] = label_quarter(data_day)


```

```python
# Tạo các DataFrame mới chỉ bao gồm 'Close' và 'quarter_label', giữ nguyên 'index'
df_day = data_day[['Close', 'quarter_label']].copy()

df_min = data_min[['Close', 'quarter_label']].copy()
df_min['quarter_label'] = label_quarter(data_min)

df_hour = data_hour[['Close', 'quarter_label']].copy()
df_hour['quarter_label'] = label_quarter(data_hour)

# Đổi tên cột 'period' trong financial_statements để khớp với 'quarter_label'
financial_statements = financial_statements.rename(columns={'period': 'quarter_label'})

# Ghép nối (merge) các DataFrame dựa trên 'quarter_label', giữ nguyên index của df gốc
merged_day = pd.merge(df_day, financial_statements, on='quarter_label', how='left', left_index=True)
merged_min = pd.merge(df_min, financial_statements, on='quarter_label', how='left', left_index=True)
merged_hour = pd.merge(df_hour, financial_statements, on='quarter_label', how='left', left_index=True)

# Hiển thị kết quả của các DataFrame mới sau khi ghép nối
print(merged_day.head())
print(merged_min.head())
print(merged_hour.head())
```
